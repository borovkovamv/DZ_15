```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
happy = pd.read_csv('Happy_2019.csv')
happy.head(15)
```

```python
happy_rank = happy.head(15)
plt.barh(happy_rank['Country or region'], happy_rank['Score'])
plt.xlabel('Score')
plt.ylabel('Country or region')
plt.title('Top 15 happiest regions')
plt.grid()
```

Вывод: Самый счастливый регион Финлядндия.

```python
happy_index = happy[list(filter(lambda x: x in ('Country or region', 'Social support', 'Healthy life expectancy', 'GDP per capita'), happy.columns))].head(15)
happy_index = happy_index.set_index(['Country or region'])
# happy_index
happy_index.plot(kind='bar', rot=45)
```

Вывод: у Люксембурга самый высокий ВВП на душу населения среди топ-15 счастливых стран, но он находится всего на 14 месте.
Самая большая социальная поддержка среди топ-15 счастливых стран у Финляндии и она находится на 1 месте.
Среди показателей ВВП на душу населения, социальная поддержка и ожидаемая продолжительность жизни наибольшее влияние оказывает социальная поддержка.

```python
happy_top50 = happy.head(50)
happy_top50
```

```python
ax = happy_top50['GDP per capita'].hist()
ax.set_title('GDP per capita distribution')
ax.set_xlabel('GDP per capita')
ax.set_ylabel('quantity')
```

Вывод: в топ-50 счастливых стран ВПП на душу населения распределн между индексом 1.3 и 1.4

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix.head()
```

```python
import seaborn as sns
```

```python
netflix_type = netflix[['type','release_year']]
netflix_type['Count'] = 1
netflix_type.head()
                                
```

```python
netflix_type_year = netflix_type.pivot_table(
                        index='release_year', 
                        columns='type', 
                        values='Count', 
                        aggfunc=sum).fillna(0).applymap(int)

netflix_type_year
```

```python
ax = sns.heatmap(netflix_type_year)
```

Вывод: За последнее время скорее всего больше статистики по фильмам и сериалам. Netflix фильмов  снимает больше, чем ТВ шоу
